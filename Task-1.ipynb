```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# 1. Upload the image
image_path = 'DIP_Lab_task_01.png'
image = Image.open("/content/DIP_Lab_task_01.png")

# 2. Image Dimensions
width, height = image.size
print(f"Image Dimensions: Width = {width}, Height = {height}")

# 3. Frequency Distribution of Intensities (Histogram)
image_array = np.array(image)
# If the image is RGB, convert it to grayscale first
if len(image_array.shape) == 3:
    image_array = np.mean(image_array, axis=2).astype(np.uint8)

plt.hist(image_array.ravel(), bins=256, range=(0, 256))
plt.title('Frequency Distribution of Intensities')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.show()

# 4. Range of Intensities
min_intensity = np.min(image_array)
max_intensity = np.max(image_array)
print(f"Range of Intensities: Minimum = {min_intensity}, Maximum = {max_intensity}")
```
