###2. Apply Image(DIP_Lab_task_03.png) smoothing with appropriate kernel applied. Identify the noise(if any) before processing.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_path = '/content/DIP_Lab_task_03.bmp'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Identify Noise (Visual inspection and histogram analysis)
plt.hist(img.ravel(),256,[0,256]); plt.show()

# Apply Smoothing (Example using Gaussian blur)
kernel_size = (5, 5)
smoothed_img = cv2.GaussianBlur(img, kernel_size, 0)

# Display Results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1); plt.imshow(img); plt.title('Original Image')
plt.subplot(1, 2, 2); plt.imshow(smoothed_img); plt.title('Smoothed Image')
plt.show()
```
